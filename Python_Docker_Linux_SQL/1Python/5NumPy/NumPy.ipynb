### Упр. - Создание массивов NumPy (ndarray)

```python
# Из списка Python
arr1 = np.array([1, 2, 3, 4, 5])
# Массив из нулей
arr2 = np.zeros(5)
# Массив из единиц
arr3 = np.ones((2, 3))
# Диапазон значений
arr4 = np.arange(0, 10, 2)
```

### Упр. - Свойства массивов NumPy (ndarray)
Размерность (ndim): Количество осей (измерений) массива.
Форма (shape): Кортеж целых чисел, представляющий количество элементов по каждой оси.
Тип данных (dtype): Тип данных элементов массива (например, int, float, bool).

```python
arr = np.array([[1, 2, 3], [4, 5, 6]])
print("Размерность:", arr.ndim)       # Вывод: 2
print("Форма:", arr.shape)          # Вывод: (2, 3)
print("Тип данных:", arr.dtype)     # Вывод: int64 (зависит от платформы)
```

### Упр. - Арифметические операции (ndarray)
Арифметические операции применяются поэлементно.

```python
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
print(arr1 + arr2)  # Вывод: [5 7 9]
print(arr1 * 2)     # Вывод: [2 4 6]
```

### Упр. - Индексация и срезы (ndarray)
Доступ к элементам и подмассивам осуществляется аналогично спискам Python.

```python
arr = np.array([0, 1, 2, 3, 4, 5])
print(arr[0])        # Вывод: 0
print(arr[1:4])     # Вывод: [1 2 3]
print(arr[::-1])    # Вывод: [5 4 3 2 1 0] (обратный порядок)
```

### Упр. - Векторизация: ускорение вычислений с помощью NumPy
Векторизация позволяет применять операции ко всему массиву без использования циклов, что значительно повышает производительность.
Пример: Возведение всех элементов массива в квадрат.

```python
# Без векторизации (медленно)
arr = np.arange(1, 1000000)
result = []
for i in arr:
    result.append(i**2)
#print(result) - очень долго
```

```python
 
# С векторизацией (быстро)
result = arr**2
print(result)
```

### Упр. - Расчет индекса массы тела (ИМТ)

   - np.mean(),
   - np.median(),
   - np.std(),

```python
data = np.array([12, 18, 22, 15, 20, 25, 30, 28])
mean_value = np.mean(data)
median_value = np.median(data)
std_deviation = np.std(data)
variance = np.var(data)

print("Mean:", mean_value, 
      "\nMedian:", median_value, 
      "\nStandard Deviation:", round(std_deviation, 2),
     "\nVariance:", round(variance, 2))
```

```python
data_x = np.array([15, 18, 20, 22, 25, 30, 28, 35])
data_y = np.array([50, 55, 60, 65, 70, 75, 80, 85])
correlation_matrix = np.corrcoef(data_x, data_y)
correlation_coefficient = correlation_matrix[0, 1]
print("Correlation Coefficient:", round(correlation_coefficient,3))
print("\nCorrelation_matrix:\n", correlation_matrix)
```

```python
# Imoprting required libraries.
# import numpy as np
# import matplotlib.pyplot as plt
import pandas as pd
# создадим датафрейм из словаря
d = {"Hours": [2.5,	5.1, 3.2, 8.5,	
               3.5,	1.5, 9.2, 5.5,	
               8.3,	2.7, 7.7, 5.9,	
               4.5,	3.3, 1.1, 8.9,	
               2.5,	1.9, 6.1, 7.4,	
               2.7,	4.8, 3.8, 6.9,7.8],

    "Scores": [21, 47, 27, 75, 30, 20, 88,
               60, 81, 25, 85, 62, 41, 
               42, 17, 95, 30, 24, 67, 69,	
               30, 54, 35, 76, 86] }
df  = pd.DataFrame(d)
# Displayinng the first five elements of the dataframe.
df.head(10)
```

В линейной регрессии мы хотим провести прямую линию к таким данным. После этого мы сможем прогнозировать результаты на любое количество часов, которое захотим.  

### Шаги:  
1. Создадим собственный простой класс для реализации модели линейной регрессии
2. При инициализации класса создадим матрицу X путем добавления к массиву x столбца, состоящего из единиц. Это нужно для получения коэффициента, определяющего пересечение с осью y
3. Также при инициализации экземпляра класса методом обратных матриц рассчитываем коэффициенты m и b для формулы линейной регрессии
`y = mx + b`
4. Создаем метод `predict` для предсказания значений на основании имеющихся коэффициентов.

```python
class MyLinearRegression:
    
    def __init__(self, x, y):
        self.X = np.vstack([x, np.ones(len(x))]).T
        self.m, self.b = np.linalg.inv(self.X.T @ self.X) @ self.X.T @ y

    def predict(self, x):
        return self.m * x + self.b
    
```

```python
# Инициализируем экземпляр класса и строим график с линией регрессии
# данные из датафрема df
model = MyLinearRegression(df['Hours'], df['Scores'])

plt.scatter(df['Hours'], df['Scores'], color='red', alpha = 0.6, edgecolors='black')
plt.plot(df['Hours'], model.predict(x), color='green', linewidth = 3)
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.title('Линейная регрессия')
plt.grid()
plt.show()
```

```python
 # Проверяем работу модели
pred_x = [20, 100, 200]
for x in pred_x:
    print("За {} часов значение оценок будет\
: {}".format(x, model.predict(x)))
```
