# Упражнения Beautiful Soup
## Упр 1
1. Используя Beautiful Soup, извлеките все заголовки новостей с главной страницы новостного сайта


```python
# сохраним наш URL в переменную:
url = 'https://lenta.ru/rubrics/russia/'
# отправим GET()-запрос на сайт и сохраним полученное в переменную:
response = requests.get(url)
html_content = response.text
# Проверим подключение:
print(response.status_code)
```

Код вернул нам статус код '200', значит это, что мы успешно подключены и все в полном порядке

Теперь воспользуемся функцией поиска в BeautifulSoup4 чтобы  извлечь заголовки новостей и ссылки на статьи с главной страницы новостного сайта. 

```python
# Создаем объект BeautifulSoup
soup = BeautifulSoup(html_content,'html.parser')
#Ищем теги с заглавными заголовками и сохр в список
links = soup.find_all(['h2','h3'])
news = soup.find_all('h4', class_='entry-title')
```

```python
# выводим заголовки из списка links
if links:
    for link in links:
        print(link.get_text(strip=True))
else:
    print("Заголовки не найдены.")
```

# Упражнения Beautiful Soup
## Упр 2
2. Найдите на странице интернет-магазина все товары определенной категории и выведите их названия и цены.
используеем демо-сайт. [`Внимание! Это демонстрационный веб-сайт для веб-скрейпинга. Цены и рейтинги здесь были распределены случайным образом и не имеют реального значения.`](https://git.astondevs.ru/laboratory/laba-data-analysis/laba_da_streams/stream10/s.makarov/-/blob/main/module2/_python/Парсинг%20и%20обработка%20веб-данных/Парсинг.ipynb?ref_type=heads)

```python
# сохраним наш URL в переменную:
url = 'https://books.toscrape.com/catalogue/category/books/thriller_37/index.html'

# отправим GET()-запрос на сайт и сохраним полученное в переменную:
response=requests.get(url)
html_content=response.text
# проверяем ответ на запрос
print(response.status_code)
```

```python
# Создаем объект BeautifulSoup
soup=BeautifulSoup(html_content,'html.parser')
```

```python
# сохраним в переменную то что ищем
class_url = 'col-xs-6 col-sm-4 col-md-3 col-lg-3'
#Ищем теги из class_url и сохр в список
products = soup.find_all(class_=class_url)
products
```

```python
# объявляем переменную в которую запишем список книг
books_list = []
for product in products:
    # Извлечение имени товара и убираем лишние символы с обеих сторон строки
    product_name = product.find('h3').text.split(".")[0].strip()
    
    # Извлечение цены товара очищаем цену от лишних символов Â 
    product_price = product.find(class_ ="price_color").text.replace("Â", "")
    
    # Добавление данных о товаре в список
    books_list.append({"name": product_name, "price": product_price})

# Вывод результатов
for item in books_list:
    print(f'Цена: {item['price']} книга: "{item['name']}"')
```

```python
# сохраняем в переменные адрес и тэг поиска
url = 'https://books.toscrape.com/catalogue/category/books/thriller_37/index.html'
class_url = 'col-xs-6 col-sm-4 col-md-3 col-lg-3'
```

```python
# Примеры - Этические аспекты парсинга 
# import time
# Установка задержки между запросами
time.sleep(5)
# Проверка файла robots.txt
from urllib.robotparser import RobotFileParser
rp = RobotFileParser()
rp.set_url(url)
rp.read()
if rp.can_fetch("*", url):
    print('Парсинг разрешен!')
else:
    print('Парсинг запрещен!!!')
```

### Упражнения - Обработка и сохранение данных
Напишите скрипт, который извлекает данные с веб-страницы, очищает их от лишних символов и сохраняет в файл CSV.

Модифицируйте скрипт, чтобы он сохранял данные в формате JSON.

добавлены этические аспекты

```python
try:
    time.sleep(5)
    # Проверка файла robots.txt
    from urllib.robotparser import RobotFileParser
    rp = RobotFileParser()
    rp.set_url(url)
    rp.read()
    if rp.can_fetch("*", url):
        print('Парсинг разрешен!')
    else:
        print('Парсинг запрещен!')
    response = requests.get(url) #ранее объявленная переменная с url
       # Проверка статуса ответа
    print('Cтатус ответа на GET-запрос: ', response.status_code)
    # 200 означает успех
    html_content = response.text
    soup = BeautifulSoup(html_content, 'html.parser')
    links = soup.find_all(class_ =class_url) #ранее объявленная переменная с тэгом
    books = []
    if not links:
        print("На странице книги не найдены")
    else:
        for link in links:
            (books.append(link.find('h3').text.split(".")[0]                         
                          .strip() + ' ;'
                          + link.find(class_ ="price_color")
                          .text.replace("Â", "")))
             
        with open ('books.csv', 'w'
                   , newline='', encoding='utf-8') as csvfile:
                
            writer = csv.writer(csvfile)
            for book in books:
                writer.writerow([book])
            print('Файл books.csv сохранен!')
            
        with open('books.json', 'w', encoding='utf-8') as jsonfile:
            
            json.dump(books, jsonfile, ensure_ascii=False, indent=4)
            print('Файл books.json сохранен!')
            
except requests.exceptions.RequestException as e:
    print(f"Ошибка при запросе к сайту: {e}")
    
except Exception as e:
    print(f"Неизвестная ошибка: {e}")
    
finally:
    print("Скрипт завершен!")
```
