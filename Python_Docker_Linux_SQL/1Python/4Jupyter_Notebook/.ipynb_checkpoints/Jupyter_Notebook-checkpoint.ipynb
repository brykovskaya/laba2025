### Упр. - Создание простого графика с помощью Matplotlib
1. Создайте новую ячейку кода.
1. Введите следующий код:
1. Запустите ячейку кода. Jupyter Notebook отобразит график функции y = x^2.

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.title("График функции синуса")
plt.show()
```

### Упр. - Базовые операции с числами в Python

```python
import matplotlib.pyplot as plt
import numpy as np
# Генерируем данные для графика
x = np.linspace(0, 10, 50)
y = x**2
# Строим график
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("x^2")
plt.title("График функции y = x^2")
plt.show() 
```

### Упр. - Визуализация данных
Создайте два списка:
- месяцы (с названиями месяцев с января по декабрь)
- температура (с примерными средними температурами для каждого месяца в вашем городе).
- Используйте библиотеку Matplotlib, чтобы построить график, отображающий изменение температуры в течение года.
- Подпишите оси графика ("Месяцы" и "Температура, °C").
- Добавьте заголовок "Средняя температура в течение года".

```python
months = ['янв', 'фев', 'март', 
         'март', 'апр', 'май', 
         'июнь', 'июль', 'авг', 
         'сент', 'окт', 'нояб' 
        ]

temps = [-10, -15, -3,
         5, 10, 18,
         24, 29, 30,
         18, 4, -1
        ]
```

```python
# import matplotlib.pyplot as plt
# зададим размеры отдельного графика (лучше указывать в начале кода)
plt.figure(figsize = (8,5))

plt.plot(months, temps, 
         color='salmon', 
         linestyle = 'dotted', 
         linewidth = 2, 
         label = 'avg t, °C')
plt.xlabel('Месяцы', fontsize = 16)
plt.ylabel('Температура, °C', fontsize = 16)
plt.title('Средняя температура в течение года', fontsize = 18)
# выведем легенду с указанием места на графике и размера шрифта
plt.legend(loc = 'upper left', prop = {'size': 12})
# plt.grid(axis='x')
plt.grid(True)
# добавим пределы шкал
ax = plt.gca()
ax.set_ylim([-20, 40])
plt.show()
```

### Дополнительные задания
Поэкспериментируйте с различными типами графиков в Matplotlib: 
1. bar chart, 
1. scatter plot, 
1. histogram.


```python
# 1 bar chart,
# import matplotlib.pyplot as plt
fig, ax = plt.subplots()

fruits = ['яблоко', 'слива', 'вишня', 'мандарин']
weight = [350, 80, 15, 200]
bar_labels = ['red', 'blue', '_red', 'orange']
bar_colors = ['tab:red', 'tab:blue', 'tab:red', 'tab:orange']

ax.bar(fruits, weight, label=bar_labels, color=bar_colors)

ax.set_ylabel('Вес фрукта')
ax.set_title('Фрукты по цвету и весу')
ax.legend(title='Цвет фрукта')
plt.show()
```

```python
# 2 scatter plot
# import matplotlib.pyplot as plt
import numpy as np

np.random.seed(1500)

fig, ax = plt.subplots()
for color in ['tab:blue', 'tab:red', 'tab:green']:
    n = 50
    x, y = np.random.rand(2, n)
    scale = 100.0 * np.random.rand(n)
    ax.scatter(x, y, c=color, s=scale, label=color,
               alpha=0.8, edgecolors='black')
ax.set_title('График scatter plot')
plt.legend(loc = 'lower right', prop = {'size': 8})

ax.grid(True)
plt.show()
```

```python
# 3 histogram.
#import numpy as np

# Create a figure with a specific size (width, height)
plt.figure(figsize=(9, 6))
# Sample data for demonstration
np.random.seed(42)
data1 = np.random.normal(0, 1, 1000)
data2 = np.random.normal(2, 1, 1000)

plt.hist([data1, data2], bins=20, color=['skyblue', 'salmon'], edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Настройка цветов для различных ячеек в гистограммах')
plt.legend(['Data 1', 'Data 2'])
plt.show()
```
