## üë®üèª‚Äçüíª –ü—Ä–∞–∫—Ç–∏–∫–∞2 -  –°–ª—É—á–∞–π–Ω—ã–π –ª–µ—Å (Random Forest) –∫–∞–∫ –ø—Ä–∏–º–µ—Ä –±—ç–≥–≥–∏–Ω–≥–∞

### ‚úî 1. –ò–º–ø–æ—Ä—Ç: 
–ò–º–ø–æ—Ä—Ç–∏—Ä—É–µ–º –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã–µ –±–∏–±–ª–∏–æ—Ç–µ–∫–∏, –≤–∫–ª—é—á–∞—è RandomForestClassifier –∏–∑ sklearn.ensemble.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

### ‚úî 2. –î–∞–Ω–Ω—ã–µ: 
–ó–∞–≥—Ä—É–∂–∞–µ–º –Ω–∞–±–æ—Ä –¥–∞–Ω–Ω—ã—Ö iris –∏ —Ä–∞–∑–¥–µ–ª—è–µ–º –µ–≥–æ –Ω–∞ –æ–±—É—á–∞—é—â—É—é –∏ —Ç–µ—Å—Ç–æ–≤—É—é –≤—ã–±–æ—Ä–∫–∏.

```python
# –ó–∞–≥—Ä—É–∑–∫–∞ –¥–∞–Ω–Ω—ã—Ö
iris = load_iris()
X = iris.data
y = iris.target

# –†–∞–∑–¥–µ–ª–µ–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö –Ω–∞ –æ–±—É—á–∞—é—â—É—é –∏ —Ç–µ—Å—Ç–æ–≤—É—é –≤—ã–±–æ—Ä–∫–∏
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

### ‚úî 3. –ú–æ–¥–µ–ª—å: 
–°–æ–∑–¥–∞–µ–º –æ–±—ä–µ–∫—Ç RandomForestClassifier —Å 100 –¥–µ—Ä–µ–≤—å—è–º–∏ (n_estimators=100) –∏ random_state=0 –¥–ª—è –≤–æ—Å–ø—Ä–æ–∏–∑–≤–æ–¥–∏–º–æ—Å—Ç–∏.

```python
# –°–æ–∑–¥–∞–Ω–∏–µ –º–æ–¥–µ–ª–∏ Random Forest
# n_estimators - –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –¥–µ—Ä–µ–≤—å–µ–≤ –≤ –ª–µ—Å—É

rf_clf = RandomForestClassifier(n_estimators=100, random_state=0)
```

### ‚úî 4. –û–±—É—á–µ–Ω–∏–µ: 
–û–±—É—á–∞–µ–º –º–æ–¥–µ–ª—å –Ω–∞ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö.

```python
# –û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏
rf_clf.fit(X_train, y_train)
```

### ‚úî 5. –ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–µ:
–ü–æ–ª—É—á–∞–µ–º –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –Ω–∞ —Ç–µ—Å—Ç–æ–≤—ã—Ö –¥–∞–Ω–Ω—ã—Ö.

```python
# –ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–µ –Ω–∞ —Ç–µ—Å—Ç–æ–≤—ã—Ö –¥–∞–Ω–Ω—ã—Ö
y_pred = rf_clf.predict(X_test)
```

### ‚úî 6. –û—Ü–µ–Ω–∫–∞: 

–í—ã—á–∏—Å–ª—è–µ–º –∏ –≤—ã–≤–æ–¥–∏–º —Ç–æ—á–Ω–æ—Å—Ç—å –º–æ–¥–µ–ª–∏ –Ω–∞ —Ç–µ—Å—Ç–æ–≤—ã—Ö –¥–∞–Ω–Ω—ã—Ö.
–í–∞–∂–Ω–æ—Å—Ç—å –ø—Ä–∏–∑–Ω–∞–∫–æ–≤:
   - –ü–æ–ª—É—á–∞–µ–º –≤–∞–∂–Ω–æ—Å—Ç—å –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –∏–∑ –æ–±—É—á–µ–Ω–Ω–æ–π –º–æ–¥–µ–ª–∏ (rf_clf.feature_importances_).

   - –°—Ç—Ä–æ–∏–º —Å—Ç–æ–ª–±—á–∞—Ç—É—é –¥–∏–∞–≥—Ä–∞–º–º—É, —á—Ç–æ–±—ã –≤–∏–∑—É–∞–ª–∏–∑–∏—Ä–æ–≤–∞—Ç—å –≤–∞–∂–Ω–æ—Å—Ç—å –∫–∞–∂–¥–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞.

```python
# –û—Ü–µ–Ω–∫–∞ —Ç–æ—á–Ω–æ—Å—Ç–∏
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}")
```

### ‚úî 7. –†–µ–∑—É–ª—å—Ç–∞—Ç:

–ö–æ–¥ –≤—ã–≤–µ–¥–µ—Ç —Ç–æ—á–Ω–æ—Å—Ç—å –º–æ–¥–µ–ª–∏ (accuracy) –Ω–∞ —Ç–µ—Å—Ç–æ–≤—ã—Ö –¥–∞–Ω–Ω—ã—Ö. –î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω–æ, –±—É–¥–µ—Ç –ø–æ—Å—Ç—Ä–æ–µ–Ω–∞ –¥–∏–∞–≥—Ä–∞–º–º–∞, –ø–æ–∫–∞–∑—ã–≤–∞—é—â–∞—è –≤–∞–∂–Ω–æ—Å—Ç—å –∫–∞–∂–¥–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞ –¥–ª—è –º–æ–¥–µ–ª–∏ Random Forest. –ß–µ–º –±–æ–ª—å—à–µ –¥–ª–∏–Ω–∞ —Å—Ç–æ–ª–±—Ü–∞, —Ç–µ–º –≤–∞–∂–Ω–µ–µ –¥–∞–Ω–Ω—ã–π –ø—Ä–∏–∑–Ω–∞–∫ –¥–ª—è –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏.

```python
# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è –≤–∞–∂–Ω–æ—Å—Ç–∏ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤

importances = rf_clf.feature_importances_
plt.figure(figsize=(8, 3))
plt.barh(iris.feature_names, importances, color='violet', edgecolor='grey', lw=2
        )
plt.xlabel("–í–∞–∂–Ω–æ—Å—Ç—å –ø—Ä–∏–∑–Ω–∞–∫–∞", fontweight='bold')
plt.ylabel("–ü—Ä–∏–∑–Ω–∞–∫", fontweight='bold')
plt.title("–í–∞–∂–Ω–æ—Å—Ç—å –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –≤ Random Forest",
         fontsize=14, fontweight='bold')
plt.show()
```
